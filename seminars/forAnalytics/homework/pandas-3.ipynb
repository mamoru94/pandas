Условие 1: 

1.1 Вывести на экран первые 5 строк:

```python
import pandas as pd
df = pd.read_csv('kc_final.csv', sep=',', encoding='Windows-1251')
```

```python
df.head()
```

1.2 Создать новый признак delta_renovated, который будет содержать разницу в годах между годом реновации дома и годом постройки дома. Если реновации дома не было, то в новом признаке поставьте 0:

```python
df['delta_renovated'] = df['yr_renovated'] - df['yr_built']
df['delta_renovated'].fillna(0, inplace=True)
```

1.3 Создайте признаки года продажи, месяца продажи:

```python
df['sale_year'] = pd.DatetimeIndex(df['date']).year
df['sale_month'] = pd.DatetimeIndex(df['date']).month
```

1.4 Удалите признаки date, zipcode, lat, long:

```python
df.drop(['date', 'zipcode', 'lat', 'long'], axis=1, inplace=True)
```


Условие 3: 

3.1 Найдите среднюю стоимость домов в зависимости от количества спален. Отсортируйте от меньшей стоимости к большей:


```python
df.groupby('bedrooms')['price'].mean().sort_values()
```

3.2 Найдите минимальную, среднюю и максимальную стоимости домов в зависимости от состояния дома:

```python
df.groupby('condition')['price'].agg(['min', 'mean', 'max'])
```


3.3 Постройте таблицу с подсчетом количества домов в данных в зависимости от вида на набережную и оценкой вида:

```python
df.pivot_table(index='waterfront', columns='view', values='id', aggfunc='count')
```


3.4 Каких домов в зависимости от этажности и количества спален больше?

```python
df.pivot_table(index='floors', columns='bedrooms', values='id', aggfunc='count')
```

3.5 Постройте таблицу с подсчетом медианной стоимости домов в данных в зависимости от состояния дома и оценки дома:

```python

df.pivot_table(index='condition', columns='grade', values='price', aggfunc='median')
```
