Условие 3: 3 задача
Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
# scatter plot для отображения зависимости между стоимостью и квадратурой жилой площади
plt.scatter(df['sqft_living'], df['price'])
plt.title('Зависимость стоимости от квадратуры жилой площади')
plt.xlabel('Площадь, кв')
plt.ylabel('Цена')
plt.show()
```

```python

# bar plot для отображения зависимости между стоимостью и количеством спален
sns.barplot(x='bedrooms', y='price', data=df)
plt.title('Зависимость стоимости от количества спален')
plt.xlabel('Количество спален')
plt.ylabel('Цена, $')
plt.show()
```

```python

# box plot для отображения зависимости между стоимостью и наличием вида на набережную
sns.boxplot(x='waterfront', y='price', data=df)
plt.title('Зависимость стоимости от наличия вида на набережную')
plt.xlabel('Наличие вида на набережную')
plt.ylabel('Цена')
plt.show()
```

```python

# violin plot для отображения зависимости между стоимостью и количеством ванных комнат
plt.figure(figsize=(20,4))
sns.boxplot(x='bathrooms', y='price', data=df)
plt.title('Зависимость стоимости от количества ванных комнат')
plt.xlabel('Количество ванных комнат')
plt.ylabel('Цена')
plt.show()
```

# выводы
Из графиков можно сделать следующие выводы:
Стоимость недвижимости имеет положительную тенденцию к росту цен с увелиением квадратурой площади и количеством спален.
Наличие вида на набережную оказывает слабое влияния на стоимость недвижимости.
Есть незначительная разница между стоимости недвижимости с кол-вом ванныx комнат в среднем.

```python
df = pd.read_csv('kc-house-data.csv')
df['date'] = pd.to_datetime(df['date'])
plt.figure(figsize=(20,40))
sns.heatmap(df[['price', 'condition', 'bedrooms', 'bathrooms', 'sqft_living', 'grade']].corr(), annot=True)
plt.title('Корреляция между характеристиками недвижимости')
plt.show()
```
